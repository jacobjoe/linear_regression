```python
df[df['bmi'] < 47]['bmi'].mean()
```

```python
def replace_bmi(col):
    if col > 47:
        return 30.53
    else:
        return col
```

We can see the smoker feature highly correlated with charges. Lets build a simple linear regression with smoker and charges

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
```

```python
X = df['smoker'].to_list()
y = df['charges'].to_list()
```

```python
X = sm.add_constant(X)
```

```python
result = sm.OLS(y, X).fit()
```

```python
print(result.summary())
```

```python
lr_model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

```python
lr_model.fit(X_train, y_train)
```

```python
pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test, pred))
```

Lets build multiple regression model with all independent features

```python
X= df.drop(['charges'], axis = 1).copy()
y = df['charges'].copy()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```
