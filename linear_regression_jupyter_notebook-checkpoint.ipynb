# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('https://raw.githubusercontent.com/jacobjoe/head_size_and_brain_weight_data/master/head_size_brain_weight_data.csv')
data.head(5)
```

```python
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
n = len(x)
mean_x = np.mean(x)
mean_y = np.mean(y)
numerator = 0
denominator = 0
```

```python
for i in range(n):
    numerator += (x[i] - mean_x) * (y[i] - mean_y)
    denominator += (x[i] - mean_x) ** 2
m = numerator / denominator
c = mean_y - (m * mean_x)
print(m, c)
```

```python
min_x = np.min(x) - 100
max_x = np.max(x) + 100
X = np.linspace(min_x, max_x, 1000)
Y = c + m * X
plt.rcParams['figure.figsize'] = (20, 10)
plt.plot(X, Y, color = '#3600ff', label = 'Regression Line')
plt.scatter(x, y, c = '#ff0006', label = 'Scatter Plot')
plt.xlabel('Head size')
plt.ylabel('Brain weight')
plt.legend()
plt.show()
```

# Finding R2 using Least square method

```python
ss_total = 0
ss_residuals = 0
for i in range(n):
    y_prediction = c + m * x[i]
    ss_total += (y[i] - mean_y) ** 2
    ss_residuals += (y[i] - y_prediction) ** 2
r2 = 1 - (ss_residuals/ss_total)
print(r2)
```

# Finding R2 using Machine learning 

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
x = x.reshape((n, 1))
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg = lin_reg.fit(x, y)
```

```python
y_prediction = lin_reg.predict(x)
```

```python
r2 = lin_reg.score(x, y)
print(r2)
```
